```python
import random

words = [
    "python", "hangman", "developer", "keyboard", "function",
    "variable", "loop", "condition", "program", "syntax",
    "debug", "compile", "execute", "input", "output",
    "algorithm", "binary", "string", "integer", "boolean"
]

print("Welcome to the Hangman Game!")
print("Guess the word!")

blank_word = random.choice(words).lower()  # ensure lowercase
blanks = ["_"] * len(blank_word)
guessed_letters = []
tries = 6

while tries > 0 and "_" in blanks:
    print("\nWord: " + " ".join(blanks))
    print(f"Wrong guesses left: {tries}")
    print("Guessed letters:", " ".join(guessed_letters))

    guess = input("Enter a letter: ").lower()

    if not guess.isalpha() or len(guess) != 1:
        print("❗ Please enter a single alphabet letter.")
        continue

    if guess in guessed_letters:
        print("⚠️ You already guessed that letter.")
        continue

    guessed_letters.append(guess)

    if guess in blank_word:
        for i, letter in enumerate(blank_word):
            if letter == guess:
                blanks[i] = guess
        print(f"✅ Good job! '{guess}' is in the word.")
    else:
        tries -= 1
        print(f"❌ Sorry, '{guess}' is not in the word.")

if "_" not in blanks:
    print("\n🎉 Congratulations! You guessed the word:", blank_word)
else:
    print("\n💀 Game Over! The word was:", blank_word)



```
